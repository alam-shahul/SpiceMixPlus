```python
import scanpy as sc
import numpy as np
import squidpy as sq
from pathlib import Path

sc.logging.print_header()
print(f"squidpy=={sq.__version__}")
```

```python
# load the pre-processed dataset
data_directory = Path("../../data/slideseqv2/")
adata = sq.datasets.slideseqv2(data_directory)
adata
```

```python
sc.pl.spatial(adata, color="cluster", spot_size=30)
```

## Hopper/Treehopper

```python
from hopper.treehopper.hoppers import hopper, treehopper, PCATreePartition # class is defined in treehoppers/hoppers.py
```

```python
h = hopper(adata.X.toarray()) # X
```

```python
k=1000
sketch = h.hop(k)
```

```python
hopper_sketch_nospatial = adata[sketch]
```

```python
sc.pl.spatial(hopper_sketch_nospatial, color="cluster", spot_size=60)
```

```python
hopper_sketch_nospatial.write("hopper_sketch_nospatial_1000.hdf5")
```

### Treehopper (no coordinates)

```python
th = treehopper(adata.X.toarray(), partition=PCATreePartition, max_partition_size=500)
th.hop(1000)
```

```python
th_sketch = th.path[:1000]
```

```python
treehopper_sketch_nospatial = adata[th_sketch]
```

```python
sc.pl.spatial(treehopper_sketch_nospatial, color="cluster", spot_size=60)
```

```python
treehopper_sketch_nospatial.write("treehopper_sketch_nospatial_1000.hdf5")
```

### Treehopper (with coordinates)

```python
adata.obsm["spatial"]
```

```python
th = treehopper(np.hstack([adata.X.toarray(), adata.obsm["spatial"]]), partition=PCATreePartition, max_partition_size=500)
th.hop(1000)
```

```python
th_sketch = th.path[:1000]
```

```python
treehopper_sketch_spatial = adata[th_sketch]
```

```python
sc.pl.spatial(treehopper_sketch_spatial, color="cluster", spot_size=60)
```

```python
treehopper_sketch_spatial.write("treehopper_sketch_spatial_1000.hdf5")
```
